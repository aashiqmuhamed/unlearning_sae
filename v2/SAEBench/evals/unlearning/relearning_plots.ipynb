# Plot by epoch

```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re

def get_params(string):
    pattern = r'multiplier(\d+)_nfeatures(\d+)_layer(\d+)_retainthres(\d+(?:\.\d+)?).pkl'
    match = re.search(pattern, string)
    if match:
        return match.groups()  # multiplier, nfeatures, layer, retainthres
    return None

def get_metrics_df(metrics_dir, epoch=None):
    df = []
    result_files = [f for f in os.listdir(metrics_dir) if f.endswith('.pkl')]
    
    for file_path in result_files:
        with open(os.path.join(metrics_dir, file_path), 'rb') as f:
            metrics = pickle.load(f)
            
        file_name = os.path.basename(file_path)
        multiplier, n_features, layer, retain_thres = get_params(file_name)
        
        row = {}
        n_se_questions = 0
        n_se_correct_questions = 0
        
        for dataset in metrics:
            if dataset == 'ablate_params':
                continue
                
            row[dataset] = metrics[dataset]['mean_correct']
            
            if dataset not in ['college_biology', 'wmdp-bio']:
                n_se_correct_questions += metrics[dataset]['total_correct']
                n_se_questions += len(metrics[dataset]['is_correct'])
        
        row['layer'] = int(layer)
        row['retain_thres'] = float(retain_thres)
        row['n_features'] = int(n_features)
        row['multiplier'] = int(multiplier)
        row['all_side_effects_mcq'] = n_se_correct_questions / n_se_questions
        
        if epoch is not None:
            row['epoch'] = epoch
            
        df.append(row)
    
    return pd.DataFrame(df)

def get_best_row_per_epoch(df):
    # Filter for rows where all_side_effects_mcq > 0.99
    filtered_df = df[df['all_side_effects_mcq'] > 0.99]
    
    # Get row with minimum wmdp-bio score
    if len(filtered_df) > 0:
        return filtered_df.loc[filtered_df['wmdp-bio'].idxmin()]
    return None

# Process all epochs
base_path = '/data/aashiq_muhamed/unlearning/SAEBench'
sae_name = 'gemma-scope-2b-pt-res_layer_3/width_16k/average_l0_142'
num_epochs = 10

# Add baseline result
baseline_row = {
    'epoch': -1,  # Using -1 to indicate baseline
    'wmdp-bio': 0.290503,
    'all_side_effects_mcq': 0.99026,
    'multiplier': 500,
    'n_features': 10,
    'retain_thres': 95.0,
    'layer': 3
}

best_rows = [baseline_row]  # Start with baseline

# Process each epoch
for epoch in range(num_epochs):
    metrics_dir = os.path.join(base_path, f'artifacts_relearning_epoch_{epoch}/unlearning/gemma-2-2b-it', 
                              sae_name, 'results/metrics')
    
    if os.path.exists(metrics_dir):
        df = get_metrics_df(metrics_dir, epoch)
        best_row = get_best_row_per_epoch(df)
        if best_row is not None:
            best_rows.append(best_row)

# Create DataFrame with all results
best_df = pd.DataFrame(best_rows)
print("\nBaseline and best results per epoch:")
print(best_df[['epoch', 'wmdp-bio', 'all_side_effects_mcq', 'multiplier', 'n_features', 'retain_thres']])

# Create the plot
plt.figure(figsize=(10, 6))

# Plot epoch points
epoch_mask = best_df['epoch'] >= 0
scatter = plt.scatter(best_df[epoch_mask]['wmdp-bio'], 
                     best_df[epoch_mask]['all_side_effects_mcq'],
                     c=best_df[epoch_mask]['epoch'],
                     cmap='viridis',
                     s=100,
                     label='Relearning epochs')

# Plot baseline point
baseline_mask = best_df['epoch'] == -1
plt.scatter(best_df[baseline_mask]['wmdp-bio'],
           best_df[baseline_mask]['all_side_effects_mcq'],
           color='red',
           marker='*',
           s=200,
           label='Baseline (no finetuning)')

# Add annotations for all points
for _, row in best_df.iterrows():
    plt.annotate(
        "Baseline" if row['epoch'] == -1 else f"Epoch {int(row['epoch'])}",
        xy=(row['wmdp-bio'], row['all_side_effects_mcq']),
        xytext=(5, 5),
        textcoords='offset points',
        fontsize=8,
        bbox=dict(facecolor='white', edgecolor='none', alpha=0.7)
    )

# Add colorbar for epochs
plt.colorbar(scatter, label='Epoch')

# Customize the plot
plt.xlabel('WMDP-Bio Accuracy')
plt.ylabel('MMLU Accuracy')
plt.title('Best Unlearning Results per Epoch')
plt.grid(True, alpha=0.3)

# Set axis limits
plt.xlim(min(best_df['wmdp-bio'])-0.05, max(best_df['wmdp-bio'])+0.05)
plt.ylim(0.99, 1.0)  # Since we're only looking at results > 0.99

# Add reference line for MMLU threshold
plt.axhline(y=0.99, color='r', linestyle='--', alpha=0.3, label='MMLU Threshold')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re

def get_params(string):
    pattern = r'multiplier(\d+)_nfeatures(\d+)_layer(\d+)_retainthres(\d+(?:\.\d+)?).pkl'
    match = re.search(pattern, string)
    if match:
        return match.groups()  # multiplier, nfeatures, layer, retainthres
    return None

def get_metrics_df(metrics_dir, epoch=None):
    df = []
    result_files = [f for f in os.listdir(metrics_dir) if f.endswith('.pkl')]
    
    for file_path in result_files:
        with open(os.path.join(metrics_dir, file_path), 'rb') as f:
            metrics = pickle.load(f)
            
        file_name = os.path.basename(file_path)
        multiplier, n_features, layer, retain_thres = get_params(file_name)
        
        row = {}
        n_se_questions = 0
        n_se_correct_questions = 0
        
        for dataset in metrics:
            if dataset == 'ablate_params':
                continue
                
            row[dataset] = metrics[dataset]['mean_correct']
            
            if dataset not in ['college_biology', 'wmdp-bio']:
                n_se_correct_questions += metrics[dataset]['total_correct']
                n_se_questions += len(metrics[dataset]['is_correct'])
        
        row['layer'] = int(layer)
        row['retain_thres'] = float(retain_thres)
        row['n_features'] = int(n_features)
        row['multiplier'] = int(multiplier)
        row['all_side_effects_mcq'] = n_se_correct_questions / n_se_questions
        
        if epoch is not None:
            row['epoch'] = epoch
            
        df.append(row)
    
    return pd.DataFrame(df)

def get_specific_row(df):
    # Filter for rows with specific parameter values
    mask = (df['multiplier'] == 500) & \
           (df['n_features'] == 10) & \
           (df['retain_thres'] == 95.0)
    
    if mask.any():
        return df[mask].iloc[0]  # Return the first matching row
    return None

# Process all epochs
base_path = '/data/aashiq_muhamed/unlearning/SAEBench'
sae_name = 'gemma-scope-2b-pt-res_layer_3/width_16k/average_l0_142'
num_epochs = 10

# Add baseline result
baseline_row = {
    'epoch': -1,  # Using -1 to indicate baseline
    'wmdp-bio': 0.290503,
    'all_side_effects_mcq': 0.99026,
    'multiplier': 500,
    'n_features': 10,
    'retain_thres': 95.0,
    'layer': 3
}

specific_rows = [baseline_row]  # Start with baseline

# Process each epoch
for epoch in range(num_epochs):
    metrics_dir = os.path.join(base_path, f'artifacts_relearning_epoch_{epoch}/unlearning/gemma-2-2b-it', 
                              sae_name, 'results/metrics')
    
    if os.path.exists(metrics_dir):
        df = get_metrics_df(metrics_dir, epoch)
        specific_row = get_specific_row(df)
        if specific_row is not None:
            specific_rows.append(specific_row)

# Create DataFrame with all results
results_df = pd.DataFrame(specific_rows)
print("\nBaseline and specific parameter results per epoch:")
print(results_df[['epoch', 'wmdp-bio', 'all_side_effects_mcq', 'multiplier', 'n_features', 'retain_thres']])

# Create the plot
plt.figure(figsize=(10, 6))

# Plot epoch points
epoch_mask = results_df['epoch'] >= 0
scatter = plt.scatter(results_df[epoch_mask]['wmdp-bio'], 
                     results_df[epoch_mask]['all_side_effects_mcq'],
                     c=results_df[epoch_mask]['epoch'],
                     cmap='viridis',
                     s=100,
                     label='Relearning epochs')

# Plot baseline point
baseline_mask = results_df['epoch'] == -1
plt.scatter(results_df[baseline_mask]['wmdp-bio'],
           results_df[baseline_mask]['all_side_effects_mcq'],
           color='red',
           marker='*',
           s=200,
           label='Baseline (no finetuning)')

# Add annotations for all points
for _, row in results_df.iterrows():
    plt.annotate(
        "Baseline" if row['epoch'] == -1 else f"Epoch {int(row['epoch'])}",
        xy=(row['wmdp-bio'], row['all_side_effects_mcq']),
        xytext=(5, 5),
        textcoords='offset points',
        fontsize=8,
        bbox=dict(facecolor='white', edgecolor='none', alpha=0.7)
    )

# Add colorbar for epochs
plt.colorbar(scatter, label='Epoch')

# Customize the plot
plt.xlabel('WMDP-Bio Accuracy')
plt.ylabel('MMLU Accuracy')
plt.title('Results per Epoch (Fixed Parameters)')
plt.grid(True, alpha=0.3)

# Set axis limits with some padding
plt.xlim(min(results_df['wmdp-bio'])-0.05, max(results_df['wmdp-bio'])+0.05)
plt.ylim(min(results_df['all_side_effects_mcq'])-0.01, max(results_df['all_side_effects_mcq'])+0.01)

plt.legend()
plt.tight_layout()
plt.show()
```
